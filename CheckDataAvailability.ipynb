# Check Data Availabity on CMIP6 Data portal and Google Could

```python
## Here input the variables 
varids = ['chldiat','chldiatos']

## Here input the experiments
expids = ['piControl']
```

### CMIP6 data portal
This part of code helps you to check which model has the variables that you are looking for on CMIP6 data portal.
https://esgf-node.llnl.gov/search/cmip6/

```python
from requests import get
```

```python
url0 = 'http://esgf-node.llnl.gov/esg-search/search/?offset=0&limit=10&type=Dataset&replica=false&latest=true&project=CMIP6&variable_id='
url1 = '&frequency=mon&experiment_id='
url2 = '&facets=mip_era%2Cactivity_id%2Cmodel_cohort%2Cproduct%2Csource_id%2Cinstitution_id%2Csource_type%2Cnominal_resolution%2Cexperiment_id%2Csub_experiment_id%2Cvariant_label%2Cgrid_label%2Ctable_id%2Cfrequency%2Crealm%2Cvariable_id%2Ccf_standard_name%2Cdata_node&format=application%2Fsolr%2Bjson'

for expid in expids:
    print("\nExperiment: {}".format(expid))
    for varid in varids:
        print('\n{}:'.format(varid))
        url = url0 + varid + url1 + expid + url2
        res = get(url)
        #print(res)
        if res.status_code == 200:
            #print('ok')
            data = res.json()
            modeldic = data['facet_counts']['facet_fields']['source_id']
            n = 0
            for item in modeldic:
                if isinstance(item,str):
                    print(item)
                    n = n + 1
            print("Total: {}".format(n))
```

```python
from intake import open_catalog
url = "https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/master.yaml"
col = open_catalog(url).climate.cmip6_gcs()
```

```python
dcat = col.search(variable_id = varids,
                  experiment_id = expids)
dcat.df
```
